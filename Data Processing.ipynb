First, use the regular season results to aggregate seasonal win, losses and totals for all the different columns for each team per season. To do this, first create a unique column for the season and winning team combination and likewise for the losing team. This gives us a value to pivot results on. 

```python
RSR["SeasonWinTeam"] = RSR["Season"].map(str)+RSR["Wteam"].map(str)
```

```python
RSR["SeasonLoseTeam"] = RSR["Season"].map(str)+RSR["Lteam"].map(str)
```

```python
import numpy as np
RSRwinningteam = pd.pivot_table(RSR, index=["SeasonWinTeam"], values=["Wscore"],  aggfunc=len)
RSRlosingteam = pd.pivot_table(RSR, index=["SeasonLoseTeam"], values=["Lscore"],  aggfunc=len)
#['Wscore', "Wfgm", "Wfga", 'Wfgm3', 'Wfga3', 'Wftm', 'Wfta','Wor', 'Wdr', 'Wast','Wto', 'Wstl', 'Wblk', 'Wpf']
```

```python
RSRwinningteamstats = pd.pivot_table(RSR, index=["SeasonWinTeam"], values=['Wscore', "Wfgm", "Wfga", 'Wfgm3', 'Wfga3', 'Wftm', 'Wfta','Wor', 'Wdr', 'Wast','Wto', 'Wstl', 'Wblk', 'Wpf', 'Lscore'],  aggfunc=np.sum)
RSRlosingteamstats = pd.pivot_table(RSR, index=["SeasonLoseTeam"], values=['Lscore', "Lfgm", "Lfga", 'Lfgm3', 'Lfga3', 'Lftm', 'Lfta','Lor', 'Ldr', 'Last','Lto', 'Lstl', 'Lblk', 'Lpf', 'Wscore'],  aggfunc=np.sum)
```

```python
RegSeasStats = {}
RSRwinningteam = RSRwinningteam.unstack()
RSRwinningteam = RSRwinningteam.unstack(level=-1)
RSRlosingteam = RSRlosingteam.unstack()
RSRlosingteam = RSRlosingteam.unstack(level=-1)
```

```python
for i in RSRwinningteam:
    if i not in RegSeasStats.keys():
        RegSeasStats[i] = {'Wins':RSRwinningteam[i][0]}
```

```python
for i in RSRlosingteam:
    if i not in RegSeasStats.keys():
        RegSeasStats[i] = {'Losses':RSRlosingteam[i][0]}
    else:
        RegSeasStats[i]["losses"] = RSRlosingteam[i][0]
```

```python
RSRwinningteamstats = RSRwinningteamstats.unstack()
RSRwinningteamstats = RSRwinningteamstats.unstack(level=-1)
```

```python
for i in RSRwinningteamstats:
    n = 0 
    for p in RSRwinningteamstats[i]:
        if n == 0:
            RegSeasStats[i]["points_against"] = p
        elif n == 1:
            RegSeasStats[i]["ast"] = p
        elif n==2:
            RegSeasStats[i]["blk"] = p
        elif n==3:
            RegSeasStats[i]["dr"] = p
        elif n==4:
            RegSeasStats[i]["fga"] = p
        elif n==5:
            RegSeasStats[i]["fga3"] = p 
        elif n==6:
            RegSeasStats[i]["fgm"] = p  
        elif n==7:
            RegSeasStats[i]["fgm3"] = p     
        elif n==8:
            RegSeasStats[i]["fta"] = p 
        elif n==9:
            RegSeasStats[i]["ftm"] = p 
        elif n==10:
            RegSeasStats[i]["or"] = p 
        elif n==11:
            RegSeasStats[i]["pf"] = p 
        elif n==12:
            RegSeasStats[i]["points_for"] = p 
        elif n==13:
            RegSeasStats[i]["stl"] = p 
        elif n==14:
            RegSeasStats[i]["to"] = p 
        n += 1
```

```python
RSRlosingteamstats = RSRlosingteamstats.unstack()
RSRlosingteamstats = RSRlosingteamstats.unstack(level=-1)
```

```python
for i in RSRlosingteamstats:
    n = 0 
    for p in RSRlosingteamstats[i]:
        if n == 0:
            try: 
                RegSeasStats[i]["ast"] += p
            except:
                RegSeasStats[i]["ast"] = p
        elif n == 1:
            try:
                RegSeasStats[i]["blk"] += p
            except:
                RegSeasStats[i]["blk"] = p 
        elif n==2:
            try:
                RegSeasStats[i]["dr"] += p
            except:
                RegSeasStats[i]["dr"] = p
        elif n==3:
            try:
                RegSeasStats[i]["fga"] += p
            except:
                RegSeasStats[i]["fga"] = p
        elif n==4:
            try:
                RegSeasStats[i]["fga3"] += p
            except:
                RegSeasStats[i]["fga3"] = p
        elif n==5:
            try:
                RegSeasStats[i]["fgm"] += p
            except:
                RegSeasStats[i]["fgm"] = p 
        elif n==6:
            try:
                RegSeasStats[i]["fgm3"] += p
            except:
                RegSeasStats[i]["fgm3"] = p   
        elif n==7:   
            try:
                RegSeasStats[i]["fta"] += p
            except:
                RegSeasStats[i]["fta"] = p 
        elif n==8:
            try:
                RegSeasStats[i]["ftm"] += p
            except:
                RegSeasStats[i]["ftm"] = p 
        elif n==9:
            try:
                RegSeasStats[i]["or"] += p
            except:
                RegSeasStats[i]["or"] = p 
        elif n==10: 
            try:
                RegSeasStats[i]["pf"] += p
            except:
                RegSeasStats[i]["pf"] = p 
        elif n==11: 
            try:
                RegSeasStats[i]["points_for"] += p
            except:
                RegSeasStats[i]["points_for"] = p 
        elif n==12:
            try:
                RegSeasStats[i]["stl"] += p
            except:
                RegSeasStats[i]["stl"] = p 
        elif n==13:
            try:
                RegSeasStats[i]["to"] += p
            except:
                RegSeasStats[i]["to"] = p 
        elif n == 14:
            try:
                RegSeasStats[i]["points_against"] += p
            except:
                RegSeasStats[i]["points_against"] = p
        n += 1
```

```python

MasterDF2["Round"] = MasterDF2['TWins'].map(lambda x: 1 if x < 1 else x+1 )
```

```python
def round_points(x):
    if x == 1:
        return 1
    elif x == 2:
        return 2
    elif x == 3:
        return 4
    elif x ==4:
        return 8
    elif x ==5:
        return 16
    elif x == 6:
        return 32
    elif x == 7:
        return 64
MasterDF2['TourneyPoints'] = MasterDF2["Round"].map(lambda x: round_points(x))
```
