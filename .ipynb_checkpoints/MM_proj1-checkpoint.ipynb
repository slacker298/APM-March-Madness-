```python
#===========================================================================#
#### Import needed Packages
#===========================================================================#
import pandas as pd #Pandas-- data framework 
import glob # for grabbing from file system
import numpy as np
import statistics
#===========================================================================#
```

```python
## FOR THIS TO WORK you have to have all of the csv's in the path directory folder (and nothing else)
path =r'C:/Users/BLanier/Documents/MSBA Fall 2017/Advanced Pred Modeling/MARCH MADNESS APM PROJECT' # use your path
all_files = glob.glob(path + "/*.csv")
df = pd.DataFrame()
filename = path
filename_check = ''
desired_filename = ''
stop = True
# filename_list = []
df_name_list = []


## Create a function that will write all the df read functions for me 
for file_ in all_files:
    for i in range(len(file_)): #For all of the files in the folder...
        filename_check += file_[i] #Build up the filename_check string
        if filename_check == filename: #Change boolean when string is full of crap
            stop = False
        if stop == False: # When string is full of crap...
            desired_filename += file_[i] # Pull the desired filename
    stop = True # Reset boolean 
    
    # Some careful fuckery of the filename to get desired output
    desired_filename = desired_filename[2:]
    df_name = desired_filename[:-4]
    df_name = 'df_'+df_name
    func_name = df_name + "=" + "pd.read_csv('"+desired_filename+"')"
    
    desired_filename = '' # Reset filename
    filename_check = '' # Reset filename_check
    df_name_list.append(df_name) # Create list of df names
    
    print(func_name) # Print out functions to copy paste into next cell 
```
